## Feature Engineering

Pada tahap ini dilakukan feature engineering perhitungan rata-rata ratings dari seluruh rating pada masing-masing coffe store. Perhitungan ini sudah dilakukan dengan mengabaikan NaN value. Jadi, ketika terdapat nilai NaN pada fitur rating tersebut, maka fitur tersebut tidak diikutkan untuk perhitungan rata-rata rating pada masing-masing coffee store. Misalkan pada 4 fitur rating, terdapat nilai NaN pada fitur YummiAdvisorRating, maka perhitungan rata-rata hanya menghitung rating GoogleRating, GoFoodRating, GrabFoodRating.

```python
sum_ratings = df[["GoogleRating", "GoFoodRating", "GrabFoodRating", "YummyAdvisorRating"]].sum(axis=1)
count_ratings = df[["GoogleRating", "GoFoodRating", "GrabFoodRating", "YummyAdvisorRating"]].notna().sum(axis=1)
df["AverageRatings"] = sum_ratings / count_ratings
df
```

### Rata-rata rating pada masing-masing kategori outlet

```python
avgRating_by_category = df.groupby("OutletCategory")["AverageRatings"].mean()
avgRating_by_category = avgRating_by_category.sort_values(ascending=False)
avgRating_by_category.plot(kind="bar", figsize=(8,5))
plt.title("Rata-rata rating pada Kategori Outlet")
plt.ylabel("Rata-rata rating")
plt.xlabel("Kategori Outlet")
plt.xticks(rotation=45, ha="right")
plt.ylim(4)
plt.tight_layout()
plt.show()
```

Kategori outlet Univesity memiliki rata-rata rating terbesar dibanding kategori outlet lainnya. Hal ini dapat menunjukkan investor terkait kategori outlet yang mendapat kepuasan pelanggan tertinggi.

### Rata-rata rating per Provinsi di Indonesia

```python
avgRating_by_province = df.groupby("Province")["AverageRatings"].mean()
avgRating_by_province = avgRating_by_province.sort_values(ascending=False)
avgRating_by_province.plot(kind="bar", figsize=(8,5))
plt.title("Rata-rata rating per Provinsi")
plt.ylabel("Rata-rata rating")
plt.xlabel("Provinsi")
plt.xticks(rotation=45, ha="right")
plt.ylim(4)
plt.tight_layout()
plt.show()
```
